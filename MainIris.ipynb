```python
def seperate_columns(col=None):
    sepal_length = satir_ad(cicek_data)[0]
    sepal_width = satir_ad(cicek_data)[1]
    petal_length = satir_ad(cicek_data)[2]
    petal_width = satir_ad(cicek_data)[3]
    iris = satir_ad(cicek_data)[4]
    
    if col == 0:
        return sepal_length
    
    elif col == 1:
        return sepal_width
    
    elif col == 2:
        return petal_length
    
    elif col == 3:
        return petal_width
    
    elif col == 4:
        return iris
    
    elif col == None:
        return satir_ad(cicek_data)
    
seperate_columns()
```

```python
def zip_data():
    zipped_data = zip(seperate_columns(0), seperate_columns(1), seperate_columns(2), seperate_columns(3), seperate_columns(4))
    return list(zipped_data)
    
zip_data()
```

```python
def tablo_olustur():
    """
    args = satırlar
    sepal_length
    sepal_width
    petal_length
    petal_width
    iris 
    """
    tablo_duzen = '|{:<15}|{:^15}|{:^15}|{:^15}|{:>15}|'
        
    print('|{:<15}|{:^15}|{:^15}|{:^15}|{:>15}|'.format(*sutun_ad(cicek_data)))
    print(80*"-")
    for sl, sw, pl, pw, i in zip(seperate_columns(0), seperate_columns(1), seperate_columns(2), seperate_columns(3), seperate_columns(4)):
        print(tablo_duzen.format(sl, sw, pl, pw, i))
        print(80*"-")
        
    
tablo_olustur()
```

```python
def cicek_ara(sepal_l=None, sepal_w=None, petal_l=None, petal_w=None, iris=None):
    """
    sepal_l=None, 
    sepal_w, 
    petal_l=None, 
    petal_w=None, 
    iris=None
    """
    l = []
    
    for i in zip_data():
        
        if sepal_l in i or sepal_w in i or petal_l in i or petal_w in i or iris in i:
            l.append(i)
        
    return l
    
    
cicek_ara(sepal_l=5.0)
```
