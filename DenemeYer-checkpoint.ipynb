```python
import pandas as pd
import numpy as np
```

```python
bolge_data = pd.read_excel("arastirmaBolgeleri.xlsx", index_col=0)
cicek_data = pd.read_excel("irisVeriKumesi.xlsx", index_col=0)

bolge_data = bolge_data.reset_index()  # make sure indexes pair with number of rows
cicek_data = cicek_data.reset_index()
```

```python
def sutun_ad(data):
    data_bolge_columns = []
    for col in data.columns:
        data_bolge_columns.append(col)
        
    return data_bolge_columns
    
sutun_ad(bolge_data)
```

```python
def satir_ad(data):
    satir_isim = []
    
    for colName, colData in data.iteritems():
        satir_isim.append(colData.values)
        
    return satir_isim

satir_ad(bolge_data)
```

```python
sutun_ad(bolge_data)
```

```python
def seperate_columns(col=None):
    yer_kodu = satir_ad(bolge_data)[0]
    ilce = satir_ad(bolge_data)[1]
    il = satir_ad(bolge_data)[2]
    bolge_adi = satir_ad(bolge_data)[3]
    
    if col == 0:
        return yer_kodu
    
    elif col == 1:
        return ilce
    
    elif col == 2:
        return il 
    
    elif col == 3:
        return bolge_adi
    
    elif col == None:
        return satir_ad(bolge_data)
    
seperate_columns()
```
