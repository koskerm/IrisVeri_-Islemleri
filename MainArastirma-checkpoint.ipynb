```python
bolge_data = pd.read_excel("arastirmaBolgeleri.xlsx", index_col=0)

bolge_data = bolge_data.reset_index()

bolge_data.head(10)
```

```python
def seperate_columns(col=None):
    yer_kodu = satir_ad.satir_ad(bolge_data)[0]
    ilce = satir_ad.satir_ad(bolge_data)[1]
    il = satir_ad.satir_ad(bolge_data)[2]
    bolge_adi = satir_ad.satir_ad(bolge_data)[3]
    
    if col == 0:
        return yer_kodu
    
    elif col == 1:
        return ilce
    
    elif col == 2:
        return il 
    
    elif col == 3:
        return bolge_adi
    
    elif col == None:
        return satir_ad.satir_ad(bolge_data)
    
seperate_columns()
```

```python
def zip_data():
    zipped_data = zip(seperate_columns(0), seperate_columns(1), seperate_columns(2), seperate_columns(3))
    return list(zipped_data)

zip_data()[:10]
```

## Tablo

```python
def tablo_olustur():
    """
    args = satırlar
    """
    tablo_duzen = '|{:<15}|{:^15}|{:^15}|{:>15}|'
        
    print('|{:<15}|{:^15}|{:^15}|{:>15}|'.format(*sutun_ad.sutun_ad(bolge_data)))
    for yk,ilce,il,ba in zip(seperate_columns(0), seperate_columns(1), seperate_columns(2), seperate_columns(3)):
        print(80*"-")
        print(tablo_duzen.format(yk,ilce,il,ba))
        
    
tablo_olustur()
```
